```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import seaborn as sns
# File to Load (Remember to change these)
#city_data_to_load = "data/city_data.csv"
#ride_data_to_load = "data/ride_data.csv"

city_data_to_load = os.path.join("data","city_data.csv")
ride_data_to_load = os.path.join("data","ride_data.csv")


# Read the City and Ride Data
city_data=pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)



ride_data.head(2)
city_data.head(2)

```

```python
#Drop any duplicate row in city

#total_rows = city_data.count
city_data=city_data.drop_duplicates('city')
total_rows = city_data.count
# No duplicates in city found
#total_rows
city_data.head()
```

```python
ride_data.head()
```

```python
# Combine the data into a single dataset"
city_ride_df=pd.merge(city_data, ride_data, on='city',how="outer")
# Display the data table for preview
city_ride_df.head()
```

```python
#Average Fare ($) Per City,Total Number of Rides Per City,Total Number of Drivers Per City,City Type (Urban, Suburban, Rural)
#city	date	fare	ride_id	driver_count	type
#ride_byCity_df=city_ride_df.groupby('city')

#groupby_city_type = city_ride_df.groupby(["type"] + ["city"])
groupby_city_type = city_ride_df.groupby(['type','city'])
#groupby_city_type.head()

#Average Fare ($) Per City
df_fare_city_type=groupby_city_type['fare'].mean()
#df_fare_city_type.head()


#Total Number of Drivers
df_driver_city_type=groupby_city_type['driver_count'].first()
#df_rider_city_type.head()

#Total Number of Rides Per City
df_rider_city_type=groupby_city_type['ride_id'].count()
#df_rider_city_type.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# x axis based on Total number of rides per city

#x_Urban
x_Urban= df_rider_city_type['Urban']
#x_Urban.head()

#x_Suburban
x_Suburban= df_rider_city_type['Suburban']
#x_Suburban.head()

#x_Rural
x_Rural=df_rider_city_type['Rural']
#x_Rural.head()

# y axix based on Average Fare ($) Per City

#y_Urban
y_Urban=df_fare_city_type['Urban']
#y_Urban.head()

#y_Suburban
y_Suburban=df_fare_city_type['Suburban']
#y_Suburban.head()

#y_Rural
y_Rural=df_fare_city_type['Rural']
#y_Rural.head()

# Size of bubbles based on Total Number of Drivers Per City
#s_Urban
s_Urban=df_rider_city_type['Urban']
#s_Urban.head()  

#s_Suburban
s_Suburban=df_rider_city_type['Suburban']
#s_Suburban.head()

#s_Rural                                                        
s_Rural=df_rider_city_type['Rural']                                                        
#s_Rural.head()                                                      
                                                        
                                                   
                                                        
                                    


```

```python
y_max=[max(y_Urban),max(y_Suburban),max(y_Rural)]
y_min=[min(y_Urban),min(y_Suburban),min(y_Rural)]
y_max
x_max=[max(x_Urban),max(x_Suburban),max(x_Rural)]
x_min=[min(x_Urban),min(x_Suburban),min(x_Rural)]
x_max
```

```python
plt.scatter(x_Urban,y_Urban, marker="o", facecolors="Coral", edgecolors="black",
            s=s_Urban*10, alpha=0.75, linewidth=1,label="Urban")
plt.scatter(x_Suburban,y_Suburban, marker="o", facecolors="SkyBlue", edgecolors="black",
            s=s_Suburban*10, alpha=0.75, linewidth=1,label="SubUrban")
plt.scatter(x_Rural,y_Rural, marker="o", facecolors="Gold", edgecolors="black",
            s=s_Rural*10, alpha=0.75, linewidth=1,label="Rural")
# Incorporate the other graph properties
#Plot Titles, Axes Labels, Legend Labels, Wedge Percentages, and Wedge Labels.
# Set the upper and lower limits of our y axis
plt.ylim(min(y_min)-5,max(y_max)+5)
plt.xlim(min(x_min)-5,max(x_max)+5)
plt.ylim(min(y_min)-5,max(y_max)+5)
plt.xlim(min(x_min)-5,max(x_max)+5)
# legend 
plt.legend(loc='upper right')
plt.titles='Pyber Ride Sharing Data (2016)'
plt.xlabel='Total Number of Rides (Per City)'
plt.ylabel='Average Fare ($)'
plt.grid()
plt.savefig("Images/Pyber Ride Sharing.png")
plt.show()
```
